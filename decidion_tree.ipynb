```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
import pandas as pd

data = pd.read_csv(r'C:\D_Drive\WorkSpace\ds_and_ml\machine_learning\datasets\titanic\train.csv')
data.head()
```

```python
input_train = data.drop(['PassengerId', 'Survived', 'Name', 'SibSp', 'Parch','Ticket', 'Cabin', 'Embarked'], axis='columns')
target_train = data[['Survived', 'Embarked']]
input_train.head()
```

```python
print(input_train['Age'].skew())

# Calculate mean, median, and mode of 'Age' column
mean_age = input_train['Age'].mean()
median_age = input_train['Age'].median()
mode_age = input_train['Age'].mode()
print(f"Mean Age: {mean_age}, Median Age: {median_age}, Mode Age: {mode_age}, modesize: {mode_age.size}")

# Determine if we should use mean or median to replace NaN values in 'Age'
# If skewness is less than -1 or greater than 1, use median; otherwise

input_train['Age'] = input_train['Age'].fillna(median_age if input_train['Age'].skew() < -1 or input_train['Age'].skew() > 1 else mean_age)
input_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Prepare input variable
le_train_sex = LabelEncoder()
input_train['Sex_n'] = le_train_sex.fit_transform(input_train['Sex'])
input_train_n = input_train.drop(['Sex'], axis='columns')
input_train_n.head()
```

```python
# Prepare target variable
# 'Survived' is the target variable, and 'Embarked' is a categorical feature that needs encoding

le_train_Embarked = LabelEncoder()
target_train['Embarked_n'] = le_train_Embarked.fit_transform(target_train['Embarked'])
target_train_n = target_train.drop(['Embarked'], axis='columns')

target_train_n.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(input_train_n, target_train_n)
```

```python
test_data = pd.read_csv(r'C:\D_Drive\WorkSpace\ds_and_ml\machine_learning\datasets\titanic\test.csv')
test_data.head()
```

```python
input_test = test_data.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
input_test['Age'] = input_test['Age'].fillna(input_test['Age'].median() if input_test['Age'].skew() < -1 or input_test['Age'].skew() > 1 else input_test['Age'].mean())
input_test['Fare'] = input_test['Fare'].fillna(input_test['Fare'].median() if input_test['Age'].skew() < -1 or input_test['Age'].skew() > 1 else input_test['Fare'].mean())   
input_test.head()
```

```python
le_test_sex = LabelEncoder()
input_test['Sex_n'] = le_test_sex.fit_transform(input_test['Sex'])
input_test_n = input_test.drop(['Sex'], axis='columns')
input_test_n.head()
```
